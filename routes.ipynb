```python
import sys
sys.path.append('../navbench')

import matplotlib.pyplot as plt
import navbench as nb
import os

# paths to datasets
dpaths = [os.path.join('datasets', d) for d in os.listdir('datasets') if not d[0] == '.']
dpaths = [d for d in dpaths if os.path.isdir(d)]

_, ax = plt.subplots(len(dpaths), 2, figsize=(10, 12))
for i, dpath in enumerate(dpaths):
    db = nb.Database(dpath)

    ax[i, 0].plot(db.x, db.y)
    ax[i, 0].axis('equal')
    ax[i, 0].set_xlabel('$x$')
    ax[i, 0].set_ylabel('$y$')

    ax[i, 1].plot(db.x, db.z)
    ax[i, 1].axis('equal')
    ax[i, 1].set_xlabel('$x$')
    ax[i, 1].set_ylabel('$z$')
```
